```python
import numpy as np
t = np.arange(0, 5, 0.01)
```

```python
plt.figure()

plt.subplot(221) # 2행2열에서 1번째
plt.plot(t, t)
plt.subplot(222)
plt.plot(t, t**2)
plt.subplot(212) # 2행1열에서 2번째
plt.plot(t, t**3)

plt.show()
```
