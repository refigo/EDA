```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(10)
y = np.array([9, 8, 7, 9, 8, 3, 2, 4, 3, 4])
t, y
```

```python
type(t), type(y)
```

```python
plt.scatter(t, y, marker='>', c=y, s=100);
plt.colorbar();
```

```python
plt.bar(t, y, color='r', width=0.7);
```

```python
y1 = np.array([3, 2, 4, 3, 4, 8, 8, 7, 9, 8])
```

```python
plt.figure()
plt.bar(t-0.2, y, color='r', width=0.3, label='apple')
plt.bar(t+0.2, y1, color='y', width=0.3, label='banana')
plt.legend()
plt.xticks(t, ('1Q', '2Q', '3Q', '4Q', '1Q', '2Q', '3Q', '4Q', '1Q', '2Q'))
plt.show()
```

```python
plt.figure()
plt.barh(t, y, color='r', label='apple')
plt.barh(t, -y1, color='y', label='banana')
plt.xticks([-10, -5, 0, 5, 10], ['-10', '-5', '0', '5', '10'])
plt.show()
```

```python
label = ['Blue', 'Green', 'Red', 'Cyan', 'Magenta', 
         'Yellow', 'Black', 'White', 'Blue', 'Green']
plt.figure()
plt.pie(y, labels=label);
```

```python
data = np.random.normal(5, 3, 1000) #  5,  3,  1000

plt.plot(data);
```

```python
plt.hist(data, bins=20);
```

```python
plt.figure()
plt.hist(data, bins=10, facecolor='red', alpha=0.4) # alpha: 
plt.hist(data, bins=20, facecolor='green', alpha=0.4)
plt.show()
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)
```

```python
plt.figure()
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.boxplot((s1, s2, s3));
```
