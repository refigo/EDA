```python
def drawGraph7():
    plt.figure(figsize=(8,6))
    plt.scatter(data_result['인구수'], data_result['소계'], s=30)
    plt.xlabel('인구수')
    plt.ylabel('CCTV')
    plt.grid()
    plt.show()
drawGraph7()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], deg=1)
fp1
```

```python
f1 = np.poly1d(fp1)
f1
```

```python
f1(1000000)
```

```python
fx = np.linspace(100000, 700000, 100)
fx
```

```python
def drawGraph8():
    plt.figure(figsize=(8,6))
    plt.scatter(data_result['인구수'], data_result['소계'], s=30)
    plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
    plt.xlabel('인구수')
    plt.ylabel('CCTV')
    plt.grid()
    plt.show()
drawGraph8()
```

```python
data_result['오차'] = data_result['소계'] - f1(data_result['인구수'])
data_result.head()
```

```python
df_descending = data_result.sort_values(by='오차', ascending=False)
df_ascending = data_result.sort_values(by='오차', ascending=True)
```

```python
df_descending.head()
```

```python
df_ascending.head()
```

```python
from matplotlib.colors import ListedColormap

color_map = ['#e74c3c', '#2ecc71', '#95a5a6', '#2ecc71', '#3498db', '#3498db']
my_colormap = ListedColormap(color_map)
```

```python
def drawGraph9():
    plt.figure(figsize=(8,7))
    plt.scatter(data_result['인구수'], data_result['소계'], 
                c=data_result['오차'], cmap=my_colormap)
    plt.plot(fx, f1(fx), ls='dashed', lw=3, c='g')
    
    for n in range(5):
        plt.text(df_descending['인구수'][n] * 1.02, # 글자가 점에 겹치지 않도록 버퍼 두기
                 df_descending['소계'][n] * 0.98, 
                 df_descending.index[n])
        
        plt.text(df_ascending['인구수'][n] * 1.02, # 글자가 점에 겹치지 않도록 버퍼 두기
                 df_ascending['소계'][n] * 0.98, 
                 df_ascending.index[n])
    plt.show()
drawGraph9()
```
