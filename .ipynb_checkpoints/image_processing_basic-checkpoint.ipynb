```python
img1 = cv2.imread('./data/dog_and_pinkwink.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('./data/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2, (600, 600))

img1.shape, img2.shape
```

```python
x_offset = img1.shape[1] - img2.shape[1]
y_offset = img1.shape[0] - img2.shape[0]

x_offset, y_offset
```

```python
roi = img1[y_offset:y_offset + img2.shape[0], x_offset:x_offset + img2.shape[1]]
plt.imshow(roi);
```

```python
img2_gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img2_gray, cmap='gray');
```

```python
mask_inv = cv2.bitwise_not(img2_gray)
plt.imshow(mask_inv, cmap='gray');
```

```python
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
plt.imshow(fg);
```

```python
final_roi = cv2.bitwise_or(roi, fg)
plt.imshow(final_roi);
```

```python
img1[y_offset:img1.shape[0], x_offset:img1.shape[1]] = final_roi
plt.imshow(img1);
```
