```python
points = np.arange(-5, 5, 0.01)
points
```

```python
points.shape
```

```python
xs, ys = np.meshgrid(points, points) # 그래프에 필요한 점들
```

```python
xs
```

```python
ys
```

```python
xs.shape, ys.shape
```

```python
xs**2 + ys**2
```

```python
z = np.sqrt(xs**2 + ys**2)
z
```

```python
plt.imshow(z)
plt.colorbar()
```

```python
plt.imshow(z, cmap=plt.cm.gray)
plt.colorbar()
plt.title('$\sqrt{x^2 + y^2}$')
plt.show()
```

```python
A = np.array([1,2,3,6])
B = np.array([1,6,8,9])
```

```python
np.in1d(A, B) # A에서 봤을 때, B에서 포함관계
```

```python
A[np.in1d(A, B)]
```

```python
np.union1d(A, B)
```

```python
np.intersect1d(A, B)
```

```python
np.setdiff1d(A, B) # A에서 B를 뺀 차집합
```

```python
np.setdiff1d(B, A)
```

```python
np.setxor1d(A, B) # XOR
```

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
```

```python
x**2 + y**2
```

```python
(x + y)**2
```

```python
(x + y)**3
```

```python
sym.expand((x + y)**3) # 식 전개
```

```python
((x**2 - y**2) / (x+y))
```

```python
sym.simplify((x**2 - y**2) / (x+y)) # 약분
```

```python
sym.limit((2*x+10)/(x+1), x, sym.oo) # 극한
```

```python
sym.sin(2*x)
```

```python
sym.diff(sym.sin(2*x), x) # 미분
```

```python
3*x**2
```

```python
sym.integrate(3*x**2, x) # 적분
```
