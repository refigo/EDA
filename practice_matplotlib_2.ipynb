```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 5, 0.5)

plt.figure()
plt.plot(t, t, t, t**2, t, t**3)
plt.show()
```

```python
t = np.arange(0, 5, 0.5)

plt.figure()
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```
